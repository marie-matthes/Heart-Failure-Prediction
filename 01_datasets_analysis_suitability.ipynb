## Heart Disease Datasets | Analysis on suitability

* we found two similar looking datasets on kaggle (one with three time the rows as the other)
* cleaning bigger dataset from duplicates to see what remains

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('./data/1025_heart_disease_dataset.csv')
df_small = pd.read_csv('./data/304_heart_attack_dataset_small.csv')
```

* renaming smaller dataset so that both datasets are structured the same

```python
df_small.rename(columns={'trtbps':'trestbps', 'thalachh': 'thalach', 
                         'exng': 'exang', 'slp':'slope', 'caa': 'ca', 'thall': 'thal', 
                         'output': 'target'}, inplace=True)
```

* check for duplicates in bigger dataset, result: 2/3 are duplicates
* 302 unique rows remaining, which is almost the same as the smaller dataset to begin with

```python
duplicate_rows = df[df.duplicated()]
duplicate_rows
```

```python
df_clean = df.drop_duplicates()
```

* combine small dataset and 'cleaned dataset', then check for duplicates again, to see whether we basically have the same data in both sets

```python
frames = [df_clean, df_small]
```

```python
df_combined = pd.concat(frames)
```

* half of the combined dataset is duplicates, meaning that we basically had the same dataset twice
* reducing the number of rows we could actually use for machine learning back to 303

```python
combined_duplicates = df_combined[df_combined.duplicated()]
combined_duplicates.describe()
```

```python
combined_cleaned = df_combined.drop_duplicates()
```
